```python
import pandas as pd
```

```python
airfoil_res= pd.read_csv("airfoil/results.csv")
airfoil_res.groupby("model")[["rmse","nll_default", "nll_mean", "train_time", "total_time"]].mean()
```

```python

airfoil_res.groupby("model")[["rmse", "nll_default", "nll_mean", "total_time"]]    .agg(["mean","std"])


```
